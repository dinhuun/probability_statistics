```python
import matplotlib.pyplot as plt
import seaborn as sns

from probability_statistics.survival_analysis import Observation, estimate_survival_function, get_estimator_coordinates
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_color_codes()

figsize = (12, 8)
```

### reproducing survival analysis of AML example in wikipedia page

```python
# those that maintained treatment
maintained_1 = Observation(9, False)
maintained_2 = Observation(13, False)
maintained_3 = Observation(13, True)
maintained_4 = Observation(18, False)
maintained_5 = Observation(23, False)
maintained_6 = Observation(28, True)
maintained_7 = Observation(31, False)
maintained_8 = Observation(34, False)
maintained_9 = Observation(45, True)
maintained_10 = Observation(48, False)
maintained_11 = Observation(161, True)
```

```python
maintained = [maintained_1, maintained_2, maintained_3, maintained_4, maintained_5, maintained_6, 
              maintained_7, maintained_8, maintained_9, maintained_10, maintained_11]
```

```python
# those that did not maintain treatment
nonmaintained_12 = Observation(5, False)
nonmaintained_13 = Observation(5, False)
nonmaintained_14 = Observation(8, False)
nonmaintained_15 = Observation(8, False)
nonmaintained_16 = Observation(12, False)
nonmaintained_17 = Observation(16, True)
nonmaintained_18 = Observation(23, False)
nonmaintained_19 = Observation(27, False)
nonmaintained_20 = Observation(30, False)
nonmaintained_21 = Observation(33, False)
nonmaintained_22 = Observation(43, False)
nonmaintained_23 = Observation(45, False)
```

```python
nonmaintained = [nonmaintained_12, nonmaintained_13, nonmaintained_14, nonmaintained_15, nonmaintained_16, 
                 nonmaintained_17, nonmaintained_18, nonmaintained_19, nonmaintained_20, nonmaintained_21,
                 nonmaintained_22, nonmaintained_23]
```

```python
maintained_estimator = estimate_survival_function(maintained)
nonmaintained_estimator = estimate_survival_function(nonmaintained)
```

```python
maintained_x_s, maintained_y_s = get_estimator_coordinates(maintained_estimator)
nonmaintained_x_s, nonmaintained_y_s = get_estimator_coordinates(nonmaintained_estimator)
```

```python
plt.figure(figsize=figsize)
plt.step(maintained_x_s, maintained_y_s, color="b", label="maintained")
plt.step(nonmaintained_x_s, nonmaintained_y_s, color="g", label="nonmaintained")
plt.legend()
plt.title("Kaplanâ€“Meier graphs of survivals by those who maintained treatment and those who did not in AML dataset")
plt.xlabel("months")
plt.ylabel("survival rate")
plt.show()
```
