```python
import time

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bayes_opt import BayesianOptimization

from probability_statistics.bayesian_optimization import my_function, plot_optimizer
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_color_codes()

figsize = (12, 8)
```

```python
x_min, x_max = -5, 5
n_X = 1000
X = np.linspace(x_min, x_max, n_X).reshape(-1, 1)
Y = my_function(X)
```

```python
plt.figure(figsize=figsize)
plt.plot(X, Y)
plt.show()
```

```python
param_bounds = {'X': (x_min, x_max)}
optimizer = BayesianOptimization(my_function, param_bounds, verbose=0, random_state=0)
```

```python
n_steps = 0
n_iter = 1
while n_steps < 10:
    if n_steps == 0:
        init_points = 2
    else:
        init_points = 0
    optimizer.maximize(init_points=init_points, n_iter=n_iter, acq='ucb', kappa=5)
    plot_optimizer(optimizer, X, Y, x_min, x_max)
    n_steps += 1
    time.sleep(2)
```
